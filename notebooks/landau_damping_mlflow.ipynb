#### This notebook provides an example of how one can use VlaPy. Here, we verify that the code reproduces Landau Damping for $k=0.3$.  To initialize the example, we apply an external force to the system for a short period of time. After this time, we observe evidence of damping of the electric field at the rate specified by Landau Damping theory.

### Modify the path to contain the root directory of the repository

```python
import sys
import os
sys.path.insert(0,os.path.dirname(sys.path[0]))
```

### Import Packages

```python
from vlapy import manager
from diagnostics import landau_damping
import uuid
```

### Perform 1 run

```python
# Make temporary directory
temp_dir = os.path.join(os.getcwd(), "temp-" + str(uuid.uuid4())[-6:])
os.makedirs(temp_dir, exist_ok=True)

# Start a run
manager.start_run(
    nx=48,
    nv=512,
    nt=1000,
    tmax=100,
    nu=0.0,
    w0=1.1598,
    k0=0.3,
    a0=1e-5,
    diagnostics=landau_damping.LandauDamping(),
    name="Landau Damping",
)

# The types of plots are chosen in the external diagnostics module
```
